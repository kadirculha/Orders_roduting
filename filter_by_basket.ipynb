```python
import pandas as pd
import ast
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

def process_association_rules(df):
    # Sütunları düzenle
    df_temp = df.iloc[:, 1:]
    df_temp["edited_productno"] = df_temp["consequents"].apply(lambda x: ast.literal_eval(x.replace("frozenset({", "[").replace("})", "]")))
    df_temp["edited_productno"] = df_temp["edited_productno"].apply(lambda x: sorted([int(pro) for pro in x]))

    # Düzenlenmiş sütunu alıp işleyin
    df_ordered = df_temp["edited_productno"].value_counts().reset_index().sort_values(by="count", ascending=True).head(3000)
    df_ordered.reset_index(drop=True, inplace=True)

    # MLxtend TransactionEncoder kullanarak veriyi işleyin
    te = TransactionEncoder()
    te_ary = te.fit(df_ordered['edited_productno']).transform(df_ordered['edited_productno'])
    df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

    # Apriori algoritması ile birliktelik kurallarını çıkarın
    frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.01)

    return rules
```

```python
b1 = ReadExcelSheet(filePath="Files/üç market - tek gün - B1.xlsx",sheetIndex=0)
d1 = ReadExcelSheet(filePath="Files/üç market - tek gün - D1.xlsx", sheetIndex=0)
d2 = ReadExcelSheet(filePath="Files/üç market - tek gün - D2.xlsx", sheetIndex=0)
koz = ReadExcelSheet(filePath="Files/üç market - tek gün - KOZ.xlsx", sheetIndex=0)
nf1 = ReadExcelSheet(filePath="Files/üç market - tek gün - NF1.xlsx", sheetIndex=0)
```
